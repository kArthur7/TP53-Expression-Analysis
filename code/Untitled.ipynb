```python
import pandas as pd
df= pd.read_csv("../data/brca_synthetic_expression.csv", index_col=0)
df.head()
```

```python
sample_type= df.loc["sample_type"]
sample_type
```

```python
expr= df.drop("sample_type")
expr
```

```python
tumor_df=sample_type[sample_type=="Tumor"].index
tumor_df
```

```python
tumor_cells=expr[tumor_df]
tumor_cells
```

```python
tumor_tp53= tumor_cells.loc["TP53"]
tumor_tp53
```

```python
normal_type= sample_type[sample_type=="Normal"].index
normal_type
```

```python
normal_expr=expr[normal_type]
normal_expr
```

```python
normal_tp53=normal_expr.loc["TP53"]
normal_tp53
```

```python
tumor_tp53=pd.to_numeric(tumor_tp53, errors="coerce")
normal_tp53=pd.to_numeric(normal_tp53, errors="coerce")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
plt.boxplot([tumor_tp53,normal_tp53], labels=["Tumor","Normal"])
plt.title("TP53 Expression:Tumor Vs Normal")
plt.ylabel("Expression Level")
plt.show()
```

```python
print(tumor_tp53.dtype)
print(normal_tp53.dtype)
```

```python
from scipy.stats import ttest_ind

_, p_val = ttest_ind(tumor_tp53, normal_tp53, nan_policy="omit")
print(p_val)
```
